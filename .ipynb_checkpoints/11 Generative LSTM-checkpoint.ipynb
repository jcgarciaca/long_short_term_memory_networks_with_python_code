```python
from random import random
```

```python
def random_rectangle():
    width, height = random(), random()
    points = list()
    points.append([0., 0.])
    points.append([width, 0.])
    points.append([width, height])
    points.append([0., height])
    return points
```

```python
rect = random_rectangle()
rect.append(rect[0])
rect
```

```python
import matplotlib.pyplot as plt
from matplotlib.path import Path
from matplotlib.patches import PathPatch
```

```python
codes = [Path.MOVETO, Path.LINETO, Path.LINETO, Path.LINETO, Path.CLOSEPOLY]
path = Path(rect, codes)
```

```python
axis = plt.gca()
patch = PathPatch(path)
axis.add_patch(patch)
```

```python
def plot_rectangle(rect):
    rect.append(rect[0])
    codes = [Path.MOVETO, Path.LINETO, Path.LINETO, Path.LINETO, Path.CLOSEPOLY]
    path = Path(rect, codes)
    axis = plt.gca()
    patch = PathPatch(path)
    axis.add_patch(patch)
    axis.set_xlim(-0.1, 1.1)
    axis.set_ylim(-0.1, 1.1)
    plt.show()
```

```python
rect = random_rectangle()
plot_rectangle(rect)
```

```python
import numpy as np
```

```python
def get_samples():
    rect = random_rectangle()
    X, y = list(), list()
    for i in range(1, len(rect)):
        X.append(rect[i - 1])
        y.append(rect[i])
    X, y = np.array(X), np.array(y)
    X = X.reshape((X.shape[0], 1, 2))
    return X, y
```

```python
X, y = get_samples()
for i in range(X.shape[0]):
    print(X[i][0], '=>', y[i])
```

## Model

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
model = Sequential()
model.add(LSTM(10, input_shape = (1, 2)))
model.add(Dense(2, activation = 'linear'))
model.compile(loss = 'mae', optimizer = 'adam')
model.summary()
```

## Fit

```python
for i in range(25000):
    X, y = get_samples()
    model.fit(X, y, epochs = 1, verbose = 2, shuffle = False)
```

## Prediction

```python
def generate_rectangle(model):
    rect = list()
    last = np.array([0., 0.]).reshape((1, 1, 2))
    rect.append([[y for y in x] for x in last[0]][0])
    for i in range(3):
        yhat = model.predict(last, verbose = 0)
        last = yhat.reshape((1, 1, 2))
        rect.append([[y for y in x] for x in last[0]][0])
    return rect
```

```python
rect = generate_rectangle(model)
plot_rectangle(rect)
```
