```python
from pandas import DataFrame
```

```python
df = DataFrame()
df['t'] = [x for x in range(10)]
df
```

```python
type(df['t'])
```

```python
df['t-1'] = df['t'].shift(2)
df
```

```python
df['t-2'] = df['t-1'].shift(2)
df
```

```python
df['t-3'] = df['t-2'].shift(-2)
df
```

```python
df['t+1'] = df['t'].shift(-1)
df
```

```python
df['t'] = df['t'].apply(lambda x: float(x))
df
```
